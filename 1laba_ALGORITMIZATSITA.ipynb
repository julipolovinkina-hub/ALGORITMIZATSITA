```python
class Node:
    def __init__(self, value: int) -> None:
        self.value: int = value
        self.next: Node | None = None          # связный список


class Queue:
    def __init__(self) -> None:
        self._head: Node | None = None         # начало очереди
        self._tail: Node | None = None         # конец очереди
        self._size: int = 0

    def enqueue(self, item: int) -> None:
        node = Node(item)                      # создание узла
        if self._tail is None:                 # если очередь пуста
            self._head = self._tail = node
        else:
            self._tail.next = node             # связываем узлы
            self._tail = node                  # обновляем хвост
        self._size += 1

    def dequeue(self) -> int:
        if self.is_empty():                    # граничный случай
            raise IndexError("Нельзя удалить элемент из пустой очереди")
        value = self._head.value               # сохраняем значение
        self._head = self._head.next           # сдвигаем голову
        if self._head is None:                 # если очередь опустела
            self._tail = None
        self._size -= 1
        return value

    def is_empty(self) -> bool:
        return self._size == 0

    def size(self) -> int:
        return self._size


# Распределительная сортировка через две очереди
def radix_sort(arr: list[int]) -> list[int]:
    if not arr:                                # граничный случай: пустой массив
        return []

    negatives: list[int] = []                  # класс list (список)
    positives: list[int] = []                  # класс list (список)
    for x in arr:                              # O(n)
        if x < 0:
            negatives.append(-x)               # O(1), берем по модулю
        else:
            positives.append(x)                # O(1)

    def sort_part(numbers: list[int]) -> list[int]:
        if not numbers:                        # граничный случай
            return []
        max_val: int = max(numbers)            # O(n)
        if max_val == 0:                       # все нули
            return numbers[:]

        bit_length: int = 0                    # счётчик бит
        temp: int = max_val
        while temp > 0:                        # O(log max_val)
            bit_length += 1
            temp = temp // 2

        for bit in range(bit_length):          # O(log max_val) — проход по битам
            q0: Queue = Queue()                # очередь для бита 0
            q1: Queue = Queue()                # очередь для бита 1

            for num in numbers:                # O(n)
                if (num >> bit) & 1 == 0:      # проверка бита
                    q0.enqueue(num)            # O(1)
                else:
                    q1.enqueue(num)            # O(1)

            numbers = []                       # очистка списка
            while not q0.is_empty():           # сборка из q0
                numbers.append(q0.dequeue())   # O(1)
            while not q1.is_empty():           # сборка из q1
                numbers.append(q1.dequeue())   # O(1)

        return numbers

    sorted_pos: list[int] = sort_part(positives)   # сортировка положительных
    sorted_neg: list[int] = sort_part(negatives)   # сортировка отрицательных

    result: list[int] = []                         # итоговый список
    for i in range(len(sorted_neg) - 1, -1, -1):   # обратный проход O(n)
        result.append(-sorted_neg[i])              # O(1)
    for x in sorted_pos:                           # O(n)
        result.append(x)                           # O(1
    return result


if __name__ == "__main__":
    arr: list[int] = list(map(int, input().split()))
    print("Исходный:  ", arr)
    print("Отсортированный: ", radix_sort(arr))
```
