```python
# Лабораторная работа 3
# Задание №20
# Выполнила: Половинкина Ю. С.
# Группа: ЦИБ-251

def is_sorted(arr):                      # O(1) - объявление функции проверки
    for i in range(1, len(arr)):         # O(n) - перебор элементов массива (начиная со 2-го)
        if arr[i - 1] > arr[i]:          # O(1) - сравнение текущего элемента с предыдущим
            return False                 # O(1) - досрочный возврат False при нарушении порядка
    return True                          # O(1) - возврат True, если цикл завершился без сбоев

if __name__ == "__main__":               # O(1) - стандартная точка входа в программу
    print(is_sorted([1, 2, 3, 4]))       # O(n) - вызов функции, ожидается: True
```
