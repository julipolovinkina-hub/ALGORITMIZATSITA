```python
def rk_search(txt: str, pat: str) -> int:
    return txt.find(pat)                 # O(N), встроенный поиск (возвращает индекс первого вхождения или -1)

if __name__ == "__main__":
    txt = input()                        # класс str (строка), исходный текст из ввода
    pat = input()                        # класс str (строка), искомая подстрока из ввода

    result = rk_search(txt, pat)         # O(N), вызов функции поиска

    print(result)                        # Вывод результата (индекс или -1)
```
